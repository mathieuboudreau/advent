```python
## Imports
import os
```

```python
## Read input
lines = open('input.txt').read().splitlines()
```

```python
## Calculate
cur_sum = 0
max_sums = [0,0,0]
for line in lines:
    if line == '':
        if cur_sum > min(max_sums):
            max_sums.append(cur_sum)
            max_sums.sort()
            max_sums = max_sums[-3:]
        cur_sum = 0
    else:
        cur_sum = cur_sum+int(line)
```

```python
## Print solution

print(sum(max_sums))
```
